```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('educational_data.csv')
dataset
```

```python
data = np.asarray(dataset[['Primary_Age_Male','Primary_Age_Female']])
print(data)
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
```

```python
plt.title('Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
plt.title('Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(data))
```

```python
plt.title('Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(data, method='average'))
```

```python
#jika affinity = euclidean (default)
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5)
output = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=output, s=50, cmap='rainbow')
plt.show()
```

```python
#jika affinity = manhattan
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, affinity='manhattan', linkage='average') #tidak bisa pake ward
output = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=output, s=50, cmap='rainbow')
plt.show()
```

```python
#LATIHAN
data_latihan = pd.read_csv('educational_data.csv')
data_latihan
```

```python
#mengubah values kolom yang bertipe data string/object menjadi numerik
from sklearn.preprocessing import LabelEncoder   
for col in ['Countries and areas']:
    label_encoder = LabelEncoder()
    dataset[col] = label_encoder.fit_transform(dataset[col])
```

```python
dataset
```

```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# Inisialisasi model AgglomerativeClustering
ac_latihan = AgglomerativeClustering(n_clusters=3)

# Melakukan prediksi
output_latihan = ac_latihan.fit_predict(data_latihan)

# Menampilkan scatter plot
plt.scatter(data_latihan[:, 0], data_latihan[:, 1], c=output_latihan, s=50, cmap='rainbow')
plt.title('Dendrogram')

# Memastikan bahwa data hanya berisi nilai numerik
numerical_data = data_latihan.astype(float)  # Gantilah sesuai dengan tipe data yang sesuai

# Menghitung dan menampilkan dendrogram
dendrogram = sch.dendrogram(sch.linkage(numerical_data, method='ward'))

plt.show()
```
